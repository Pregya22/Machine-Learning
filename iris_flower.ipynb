```python
##training&testing
#splitting data on percentage basis (to print all data)

#!/usr/bin/python3


import numpy
from sklearn.datasets import load_iris

#loading all data
iris=load_iris()

#printing feature name
print(iris.feature_names)

#target names
print(iris.target_names)

#data position which you want to delete (for testing purpose)
x=[0,50,100]

only_target_training=numpy.delete(iris.target,x)
only_data_training=numpy.delete(iris.data,x,axis=0)
#target data role
print(only_target_training.size)
print(only_data_training)

#testing target
test_target=iris.target[x]
test_data=iris.data[x]
print(test_target)
print(test_data)

```

```python
## applying DecisionTree algo

#!/usr/bin/python3

import numpy
from sklearn import tree
from sklearn.datasets import load_iris

#loading all data
iris=load_iris()

#printing feature name
print(iris.feature_names)

#target names
print(iris.target_names)

#data position which you want to delete (for testing purpose)
x=[0,50,100]

only_target_training=numpy.delete(iris.target,x)
only_data_training=numpy.delete(iris.data,x,axis=0)
#target data role
print(only_target_training.size)
print(only_data_training)

#testing target
test_target=iris.target[x]
test_data=iris.data[x]
print(test_target)
print(test_data)


#calling algo

clf=tree.DecisionTreeClassifier()
trained=clf.fit(only_data_training, only_target_training)
output=trained.predict([[7,3.2,4.7,1.4]])
print(output)
```
