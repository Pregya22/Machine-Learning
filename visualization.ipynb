```python
import matplotlib.pyplot as plt

x=[2,6]
y=[4,9]

plt.xlabel("time")
plt.ylabel("speed")
plt.bar(x,y)
plt.grid(color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[2,6]
y=[4,9]
x1=[1,2]
y1=[5,6]

plt.xlabel("time")
plt.ylabel("speed")
plt.scatter(x,y, label="river",marker='x',s=150,c='y')
plt.scatter(x1,y1, label="water")
plt.legend()
plt.grid(color='r')
plt.show()
```
